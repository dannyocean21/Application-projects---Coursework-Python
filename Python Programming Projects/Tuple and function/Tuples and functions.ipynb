## (35 pts) Problem 1 — Recursive Power Function (10 pts)

```python
v = int(input("Please enter a number (base): "))
p = int(input("Please enter a number (exponent): "))

def power(v, p):
    if p == 0:
        return 1
    return v * power(v, p-1)

print(f"The {v} in the power of {p} is {power(v, p)}")
```

## (35 pts) Problem 2 — Equivalent Resistance of n Resistors in Parallel

```python
def Req(*resistances):
    if not resistances:
        return float('inf')

    reciprocal_sum = 0
    for r in resistances:
        if r != 0:
            reciprocal_sum += 1 / r

    if reciprocal_sum == 0:
        return 0
    else:
        return 1 / reciprocal_sum

# Example usage
print(Req(100,100))        
print(Req(100,330,1000))   
```

## (15 pts) Problem 4 — Remove one 'm' from tuple

```python
tuple1 = ("k", "i", "m", "m", "e", "p")

# Convert to list
temp_list = list(tuple1)
# Remove the first 'm'
for i in range(len(temp_list)):
    if temp_list[i] == 'm':
        del temp_list[i]
        break

new_tuple = tuple(temp_list)

print("Given tuple:  ", tuple1)
print("After removing:", new_tuple)
```
