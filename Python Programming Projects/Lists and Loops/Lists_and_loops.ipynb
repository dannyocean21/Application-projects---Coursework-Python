## (35 pts) Problem 1 — Sum of Two Numbers Represented as Lists
Write Python code that calculates the sum of two integers represented in two separate lists and prints the result as a list of digits.
"""

```python
def add_lists_as_numbers(list1, list2):
    num1_str = ""
    for digit in list1:
        num1_str += str(digit)
    num1 = int(num1_str)

    num2_str = ""
    for digit in list2:
        num2_str += str(digit)
    num2 = int(num2_str)

    total = num1 + num2
    total_str = str(total)
    result_list = []
    for char in total_str:
        result_list.append(int(char))

    return result_list

# Example usage
list1 = [1, 2, 3]
list2 = [7, 8, 9]
print(add_lists_as_numbers(list1, list2))

list3 = [1, 7, 8, 9]
list4 = [1, 7, 8, 9]
print(add_lists_as_numbers(list3, list4))

List1 = [1, 2, 3, 4]
List2 = [3, 4, 5, 6, 7]
print(add_lists_as_numbers(List1, List2))

List1 = [1, 2, 3, 4]
List2 = [3, 9, 5, 6, 7]
print(add_lists_as_numbers(List1, List2))
```

## (40 pts) Problem 2 — Kronecker Product (No NumPy)
Compute the Kronecker product of two matrices without using libraries.

```python
def kronecker_product(matrix1, matrix2):
    result = []
    for row1 in matrix1:
        for row2 in matrix2:
            new_row = []
            for elem1 in row1:
                for elem2 in row2:
                    new_row.append(elem1 * elem2)
            result.append(new_row)
    return result

A = [[1, 2, 3]]
B = [[0, 1, 0]]
for row in kronecker_product(A, B):
    print(row)
```

## (15 pts) Problem 3 — Strictly Increasing Check
Define a function that returns True only when n1 < n2 < n3 < n4.

```python
def increasing(n1, n2, n3, n4):
    return n1 < n2 < n3 < n4

print(increasing(43, 44, 45, 46))
print(increasing(-20, -10, -1, 0.532))
print(increasing(0, 0, 1, 2))
print(increasing(99, 3, 55, 1))
```

## (10 pts) Problem 4 — Split on Hyphens (No split method)
Given a hyphen-separated string, print each substring.

```python
def split_on_hyphens(string):
    print("Displaying each substring")
    substring = ""
    for char in string:
        if char == "-":
            print(substring)
            substring = ""
        else:
            substring += char
    print(substring)

str1 = "Emma-is-a-data-scientist"
split_on_hyphens(str1)
```
