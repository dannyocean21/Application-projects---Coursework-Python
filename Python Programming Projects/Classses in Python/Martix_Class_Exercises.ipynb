## **(50 pts) Problem 1 — Matrix Class**
Construct a class that represents matrices. The class stores the matrix in `myMatrix` and supports matrix addition using `+`.
"""

## Answer to Problem 1:

```python
class Matrix:
    def __init__(self, myMatrix):
        self.myMatrix = myMatrix

    def __add__(self, other):
        # Check if dimensions match
        if len(self.myMatrix) != len(other.myMatrix) or any(
                len(row1) != len(row2) for row1, row2 in zip(self.myMatrix, other.myMatrix)):
            raise ValueError("Error: Matrices are not the same size and cannot be added.")

        result = []
        for row_self, row_other in zip(self.myMatrix, other.myMatrix):
            result_row = [a + b for a, b in zip(row_self, row_other)]
            result.append(result_row)

        return Matrix(result)

    def __str__(self):
        return '\n'.join(' '.join(map(str, row)) for row in self.myMatrix)
```

```python
# Example usage
matrix1 = Matrix([
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
])

matrix2 = Matrix([
    [9, 8, 7],
    [6, 5, 4],
    [3, 2, 1]
])

matrix3 = Matrix([
    [1, 1, 1],
    [1, 1, 1],
    [1, 1, 1]
])

result = matrix1 + matrix2 + matrix3
print(result)
```

# **(50 pts) Problem 2 — Minimum Path Sum in Triangle**
You are given a triangle of integers. Starting at the apex, move to the base by taking steps to adjacent numbers. Find the minimum possible path sum.

```python
def find_minimum_path_sum(triangle):
    n = len(triangle)

    # Start from the second last row and move upwards
    for i in range(n - 2, -1, -1):
        for j in range(i + 1):
            triangle[i][j] += min(triangle[i + 1][j], triangle[i + 1][j + 1])

    return triangle[0][0]

# Example triangle
triangle = [
    [2],
    [5, 4],
    [1, 4, 7],
    [8, 6, 9, 6]
]

answer = find_minimum_path_sum(triangle)
print("The minimum path sum is:", answer)
```
