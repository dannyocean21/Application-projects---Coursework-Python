## (50 pts) Problem 1 — Rectangle Fit Checker
Determine whether one rectangle fits inside another (considering rotation).

```python
class Rectangle:
    def __init__(self, width, height):
        self.width = width
        self.height = height

    def fitsInside(self, other):
        # Without rotation
        if self.width <= other.width and self.height <= other.height:
            return True
        # With rotation
        if self.height <= other.width and self.width <= other.height:
            return True
        return False

while False:  # Disabled interactive input for notebook
    w1, h1 = map(int, input("Enter first rectangle (w h): ").split())
    w2, h2 = map(int, input("Enter second rectangle (w h): ").split())

    rect1 = Rectangle(w1, h1)
    rect2 = Rectangle(w2, h2)

    if rect1.fitsInside(rect2):
        print("First box fits in second box")
    elif rect2.fitsInside(rect1):
        print("Second box fits in first box")
    else:
        print("Boxes do not fit")
```

```python
rect1 = Rectangle(10, 10)
rect2 = Rectangle(14, 20)

print(rect1.fitsInside(rect2))
if rect1.fitsInside(rect2) == "True":
    print("First box fits in second box")
```

## (50 pts) Problem 2 — Histogram Equalization (4-bit, 16 gray levels)
Compute and plot the equalized histogram for a 4‑bit image with 360 total pixels.

```python
import numpy as np
import matplotlib.pyplot as plt

gray_levels = np.arange(16)
frequencies = np.array([15, 0, 0, 0, 0, 0, 0, 0, 0, 70, 110, 45, 80, 40, 0, 0])
total_pixels = 360

probabilities = frequencies / total_pixels
cdf = np.cumsum(probabilities)
cdf_min = cdf[cdf > 0].min()

equalized_levels = np.round((cdf - cdf_min) * 15 / (1 - cdf_min)).astype(int)

equalized_histogram = np.zeros_like(frequencies)
for original, equalized in zip(gray_levels, equalized_levels):
    equalized_histogram[equalized] += frequencies[original]

fig, axes = plt.subplots(1, 2, figsize=(12, 5))

axes[0].bar(gray_levels, frequencies)
axes[0].set_title("Original Histogram")
axes[0].set_xlabel("Gray Level")
axes[0].set_ylabel("Frequency")

axes[1].bar(gray_levels, equalized_histogram)
axes[1].set_title("Equalized Histogram")
axes[1].set_xlabel("Gray Level")
axes[1].set_ylabel("Frequency")

plt.tight_layout()
plt.show()
```
