## (10 pts) Problem 1 — Convert Image to Binary using Thresholding

```python
import cv2

# Load the image
image = cv2.imread("im1.jpg", cv2.IMREAD_GRAYSCALE)

# Thresholding
threshold_value = 128
_, binary_image = cv2.threshold(image, threshold_value, 255, cv2.THRESH_BINARY)

# Save the result
cv2.imwrite("binary_im1.jpg", binary_image)

# Display the result
# Uncomment these lines if running locally
cv2.imshow("Binary Image", binary_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## (40 pts) Problem 2 — Triangle Class (Area & Perimeter)

```python
import math

class Triangle:
    def __init__(self, a, b, c):
        self.a = a
        self.b = b
        self.c = c

    def perimeter(self):
        return self.a + self.b + self.c

    def area(self):
        s = self.perimeter() / 2
        return math.sqrt(s * (s - self.a) * (s - self.b) * (s - self.c))

# Example usage
triangle = Triangle(2.5, 3.0, 5.0)
triangle_perimeter = triangle.perimeter()
triangle_area = triangle.area()
print(f"Perimeter: {triangle_perimeter}")
print(f"Area: {triangle_area:.2f}")
```

## (50 pts) Problem 3 — Time Class and Addition

```python
class Time:
    def __init__(self, hours, minutes):
        self.hours = hours
        self.minutes = minutes

    def __add__(self, other):
        total_minutes = self.minutes + other.minutes
        extra_hours = total_minutes // 60
        total_hours = self.hours + other.hours + extra_hours
        total_minutes %= 60
        return Time(total_hours, total_minutes)

    def __str__(self):
        return f"{self.hours} hours and {self.minutes} minutes"

# Example usage
time_values = [
    Time(3, 20),
    Time(4, 15),
    Time(6, 35),
    Time(2, 25)
]

total_time = Time(0, 0)
for t in time_values:
    total_time += t

print("Total time:", total_time)
```
